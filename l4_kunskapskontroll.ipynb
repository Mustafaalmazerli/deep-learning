```python

from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
model = ResNet50(weights='imagenet')
img_path = 'zebra.jpg'
img_path = 'linjal.jpg'
img_path = 'tennisboll.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
x = image.img_to_array(img)
x.shape
x = np.expand_dims(x, axis=0)
x.shape
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
# 1. Take some pictures yourself and predict their class.
img_path = '/content/cop.jpg'

img = mpimg.imread(img_path)
plt.imshow(img)
plt.axis('off')  # Avaktivera axlar
plt.show()
# Ladda in bilden och förbered den för modellen
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Gör förutsägelser med modellen
preds = model.predict(x)

# Avkoda resultaten för att få en lista över de tre mest sannolika klasserna och deras sannolikheter
decoded_preds = decode_predictions(preds, top=3)[0]

# Skriv ut resultaten
print("Predicted classes:")
for i, (imagenet_id, label, score) in enumerate(decoded_preds):
    print(f"{i+1}. {label} ({score:.2f})")
# 2. Answer the question: Why is "Pretrained models" a very powerful concept?
#Överförbara kunskaper: Förtränade modeller har lärt sig att extrahera generella funktioner från stora mängder data. Dessa funktioner kan överföras till nya uppgifter och datasatser, vilket minskar behovet av att träna modeller från grunden för varje ny uppgift.
#Tidsbesparingar: Genom att använda förtränade modeller kan du spara mycket tid och resurser som annars skulle ha spenderats på att samla in och förbereda data samt träna modellen från grunden.
#Minska behovet av stora datamängder: Förtränade modeller kan användas även när du har begränsade mängder etiketterade data tillgängliga, vilket är särskilt fördelaktigt för små organisationer eller forskningsprojekt med begränsade resurser.
#Förbättrad prestanda: Förtränade modeller är ofta tränade på stora och varierade datamängder, vilket kan leda till modeller som har bättre allmän förmåga att generalisera till nya data jämfört med modeller tränade på mindre datamängder.
```

```python
# Ange en lista med sökvägar till dina bilder
img_paths = ['/content/cop.jpg', '/content/blomma.jpg', '/content/ipad.jpg']

# Ladda in den förtränade ResNet50-modellen
model = ResNet50(weights='imagenet')

# Iterera över varje bildsökväg
for img_path in img_paths:
    # Visa bilden
    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis('off')  # Avaktivera axlar
    plt.show()

    # Ladda in bilden och förbered den för modellen
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Gör förutsägelser med modellen
    preds = model.predict(x)

    # Avkoda resultaten för att få en lista över de tre mest sannolika klasserna och deras sannolikheter
    decoded_preds = decode_predictions(preds, top=3)[0]

    # Skriv ut resultaten för den aktuella bilden
    print(f"\nPredictions for {img_path}:")
    for i, (imagenet_id, label, score) in enumerate(decoded_preds):
        print(f"{i+1}. {label} ({score:.2f})")
```
